1. The MickeyMouse problem

a) Write a program that prints the numbers from 1 to 100. But for multiples of three print Mickey instead of the number and for the multiples of five print Mouse. For numbers which are multiples of both three and five print MickeyMouse

b) Put the result in a tuple and substitute Mickey with Donald and Mouse with Duck

```python
# Ex. 1.
# a)
for x in range(1,101):
    if x%3==0 and x%5==0:
        print("MickeyMouse")
    elif x%3==0:
        print("Mickey")
    elif x%5==0:
        print("Mouse") 
    else: print(x)
```

```python
# b) 
a=[x+1 for x in range(0,100)]

for x in range(0,100):
    if (x+1)%3==0 and (x+1)%5==0:
        a[x]="MickeyMouse"
    elif (x+1)%3==0:
        a[x]="Mickey"
    elif (x+1)%5==0:
        a[x]="Mouse"
    else: a[x]=x+1
        
b=(a)
c=list(b)

for x in range(0,100):
    if c[x]=="MickeyMouse":
        c[x]="DonaldDuck"     
    if c[x]=="Mickey":
        c[x]="Donald"  
    if c[x]=="Mouse":
        c[x]="Duck" 
        
print(c)       
```

2. The swap function

Write a function that swap the values of two input variables x and y (whatever the type). Try to do that also without a temporary variable

```python
# Ex. 2.

def swap0(x,y):
    temp = x
    x = y
    y = temp
    return(x,y)

def swap1(x, y):
    x,y=y,x
    return(x,y)

def swap2(x, y):
    return(y,x)

x=1
y=2

x,y=swap0(x,y)
print(x,y)

x,y=swap1(x,y)
print(x,y)

x,y=swap2(x,y)
print(x,y)
```

3\. Computing the distance

Write a function that calculates and returns the euclidean distance between two points *u* and *v*, where *u* and *v* are both 2-tuples *(x,y)*. For example, if *u=(3,0)* and *v=(0,4)*, the function should return 5

```python
# Ex. 3.

import math

def dist(u,v):
    x = u[0] - v[0]
    y = u[1] - v[1]
    return(math.sqrt(x**2 + y**2))

u = (3,0)
v = (0,4)

print(dist(u,v))
```

5\. Isolating the unique

Write a function that determines and count the unique numbers in the list *l*

```python
# Ex. 5.

l = [36, 45, 58, 3, 74, 96, 64, 45, 31, 10, 24, 19, 33, 86, 99, 18, 63, 70, 85,
 85, 63, 47, 56, 42, 70, 84, 88, 55, 20, 54, 8, 56, 51, 79, 81, 57, 37, 91,
 1, 84, 84, 36, 66, 9, 89, 50, 42, 91, 50, 95, 90, 98, 39, 16, 82, 31, 92, 41,
 45, 30, 66, 70, 34, 85, 94, 5, 3, 36, 72, 91, 84, 34, 87, 75, 53, 51, 20, 89, 51, 20]

def unique_finder(list_a):  
    list_b = [] 
    for x in list_a: 
        if x not in list_b: 
            list_b.append(x)    
    return list_b

print("There are",  len(l) - len(unique_finder(l)), " non unique numbers: ")
print("There are",  len(unique_finder(l)), "unique numbers: ")
print("The unique numbers are: ", unique_finder(l))
```

6\. Combination of functions

Write two functions - one that returns the square of a number, and one that returns the cube. Now write a third function that returns the number raised to the 6th power using the two previous functions.

```python
# Ex. 6.

def square(x):
    return x*x

def cube(x):
    return x*x*x

def sixth(x): # Combination of the two functions
    return square(cube(x))

x = 2
print(square(x))
print(cube(x))
print(sixth(x))
```

8\. Nested list comprehension

A Pythagorean triple is an integer solution to the Pythagorean theorem $a^2+b^2=c^2$. The first Pythagorean triple is (3,4,5). Find and put in a tuple all unique Pythagorean triples for the positive integers a, b and c less than 100.

```python
# Ex. 8.

a = [(i,j,k) for i in range(1,101) for j in range(i,101) for k in range(j,101) if (i*i+j*j)==(k*k)]

print("The unique Pythagorean triples for the positive integers a, b and c less than 100 are: ", a)
```

9\. Normalization

Write a function that takes a tuple of numbers and returns it with the entries normalized to one

```python
# Ex. 9. 

t = (1,2,3,4,5)

def norm(a): # I normalized considering the sum of the elements to be 1 (not as norm for vectors)
    c=0
    for i in range(len(a)): c+=(a[i])
    b=[x/c for x in a]
    return tuple(b)

y = list(norm(t))
print(sum(y))
```
